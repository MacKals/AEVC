```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show(gray)

binary = (gray <= 40)
show(binary)
```

```python
ret, binary = cv2.threshold(gray,30,255,cv2.THRESH_BINARY_INV)
show(binary)

kernel = np.ones((5,5),np.uint8)

#erosion = cv2.erode(binary,kernel,iterations = 1)
#show(erosion)

dilation = cv2.dilate(binary, kernel, iterations = 2)
show(dilation)

#opening = cv2.morphologyEx(dilation, cv2.MORPH_OPEN, kernel)
#show(opening)

img, contours, hierarchy = cv2.findContours(dilation, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

print(np.size(contours))

#print(contours)

index = 0
maxArea = 0

for i in range(np.size(contours)):
    if cv2.contourArea(contours[i])> maxArea:
        index = i
        maxArea = cv2.contourArea(contours[i])

img = cv2.drawContours(img, contours, index, 200, 10)
show(img)

contour = contours[index]

cv2.ellipse(img, cv2.fitEllipse(contour), 100, 10)

(x,y),(MA,ma),angle = cv2.fitEllipse(contour)
print('x=' + str(x) + ', y=' + str(y) + ', MA=' + str(MA) + ', ma=' + str(ma) + ', angle=' + str(angle))

rows,cols = img.shape[:2]
[vx,vy,x,y] = cv2.fitLine(contour, cv2.DIST_L2,0,0.01,0.01)
lefty = int((-x*vy/vx) + y)
righty = int(((cols-x)*vy/vx)+y)
show(cv2.line(img,(cols-1,righty),(0,lefty),100,10))

epsilon = 0.1*cv2.arcLength(contour,True)
approx = cv2.approxPolyDP(contour,epsilon,True)
show(cv2.drawContours(img, [approx], 0, (255,255,255), 3))
print("simplified contour has" + str(len(approx)) + "points")
```

```python
def pointAngle(p1, p2, p3):
        
    v1 = p2-p1
    v2 = p2-p3
        
    return np.arccos(np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2)))*180/np.pi

def anglesBetweenPoints(points):
    p = np.append([points[-1]], np.append(points, [points[0]], axis=0), axis=0)
    return [pointAngle(p[i], p[i+1], p[i+2]) for i in range(len(points))]


#print(pointAngle(np.array([1,0]), np.array([0,0]), np.array([1,0])))

#print(points)

angles = anglesBetweenPoints(points)

ind = np.argmax(angles)

straight = np.array([1 if a > 178.5 else 0 for a in angles])

def removeLoneEntriesFromBinary(a):
    
    r = np.empty_like(a)
    
    looped = np.append([a[-1]], np.append(a, [a[0]], axis=0), axis=0)
    
    for i in range(len(a)):
        if looped[i] == looped[i+2] and looped[i] != looped[i+1]:
            r[i] = looped[i]
        else:
            r[i] = looped[i+1]
    
    return r


#returns rising edges and falling edges in two seperate arrays
def edgeIndecies(a):
    
    looped = np.append(a, [a[-1]], axis=0)
    
    risingIndecies = []
    fallingIndecies = []
    
    for i in range(len(a)):
        if looped[i] > looped[i+1]:
            fallingIndecies.append(i)
        elif looped[i] < looped[i+1]:
            risingIndecies.append(i+1)
        
    return (risingIndecies, fallingIndecies)

def dist(a, b):
    return ((a[0] - b[0])**2 + (a[1] - b[1])**2)**(1/2)

def keepFourLongest(a, p):
    risingIndecies, fallingIndecies = edgeIndecies(a)
    
    if len(risingIndecies) != len(fallingIndecies):
        print('lengthFault!')
    
    if risingIndecies[0] > fallingIndecies[0]:
        fallingIndecies = np.roll(fallingIndecies, 1)
        
    distance = np.array([dist(p[risingIndecies[i], :], p[fallingIndecies[i], :]) for i in range(len(risingIndecies))])
    
    #print(distance)
    #print(risingIndecies)
    #print(fallingIndecies)
    
    
    indeciesToKeep = np.zeros((4,2))
    
    for i in range(4):
        longestIndex = np.argmax(distance)
        
        indeciesToKeep[i,0] = risingIndecies[longestIndex]
        indeciesToKeep[i,1] = fallingIndecies[longestIndex]
        
        distance[longestIndex] = 0        
    
    print(indeciesToKeep)
    indeciesToKeep = indeciesToKeep[indeciesToKeep[:,0].argsort()]
    print(indeciesToKeep)
    r = np.zeros((4, 2, 2))
        
    for i in range(4):
        r[i,0,:] = points[int(indeciesToKeep[i, 0])]
        r[i,1,:] = points[int(indeciesToKeep[i, 1])]
    
    return r

straightFiltered = removeLoneEntriesFromBinary(straight)

print(straightFiltered)

# line: [y(0), dy/dx]

def lineFromPoints(p1, p2):
    m = (p1[1]-p2[1])/(p1[0]-p2[0])
    y0 = p1[1]-m*p1[0]
    return [y0, m]

linePoints = keepFourLongest(straightFiltered, points)

lines = [lineFromPoints(p[0], p[1]) for p in linePoints]

def intersection(l1, l2):
    x = (l1[0]-l2[0])/(l2[1]-l1[1])
    y = l1[0]+l1[1]*x
    return [x,y]

corners = np.array([intersection(lines[i], lines[(i+1)%len(lines)]) for i in range(len(lines))])

def plotLine(l, xmin, xmax):
    p1 = [xmin, l[0]+xmin*l[1]]
    p2 = [xmax, l[0]+xmax*l[1]]

    plt.plot([p1[0], p2[0]], [-p1[1], -p2[1]], '-')

straightPoints = np.array([points[i] if straight[i] else [0,0] for i in range(len(points))])
plt.figure()

plt.plot(points[:,0], -points[:,1], 'ro')
#plt.plot(straightPoints[:,0], -straightPoints[:,1], 'bo') 
plt.plot(corners[:,0], -corners[:,1], 'bo')

xmin, xmax = plt.xlim()
ymin, ymax = plt.ylim()

[plotLine(l, xmin, xmax) for l in lines] 

plt.xlim((xmin, xmax))
plt.ylim((ymin, ymax))


#print(ind)
#print(np.array(angles))
#print(np.var(angles))
#print(np.mean(angles))
```
